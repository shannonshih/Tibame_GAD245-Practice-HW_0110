# 基本圖表

## 使用Matplot來製作圖表

```python
x = np.linspace(0, 10, 100)
print(x)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
# 練習題
# 請畫出tan的曲線圖
plt.plot(x, np.tan(x), '-')
```

```python
# 練習題
# 請畫出x二次方跟三次方的曲線圖
plt.plot(x, x ** 2, '-')
plt.plot(x, x ** 3, '--')
```

```python
# 練習題
# 請畫出 sigmod(x) = 1 / (1+e^-x)
plt.plot(x, 1 / (1 + np.exp(-x)), '-')
```

## 線條顏色

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
# https://www.w3schools.com/tags/ref_colornames.asp
```

## 線條樣式

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

```python
# 練習題
# 請用 紅色實線畫出sin,藍色虛線畫出cos
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-r')
plt.plot(x, np.cos(x), '--b')
```

# 圖例

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend()
```

```python
# 設定圖例的位置
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
# x, y,軸的值呈現等比
plt.axis('equal')
plt.legend(loc='upper right')
```

```python
# 練習題
# 請將上圖的圖例放在右下角位置
# 設定圖例的位置
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
# x, y,軸的值呈現等比
plt.axis('equal')
plt.legend(loc='lower right')
```

```python
# frameon -> 圖例加框(預設值為True)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc='upper right', frameon=True)
```

```python
# 圖例加框
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc='upper right', frameon=False)
```

```python
# ncol -> 圖例欄數
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc='lower center', frameon=True, ncol=2)
```

```python
# 練習題
# 請將上圖的圖例放在正上方的位置，且橫向排列
# ncol -> 圖例欄數
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc='upper center', frameon=True, ncol=2)
```

```python
# 練習題
# 請用 2x2 的圖形分別畫出 sin cos tan tanh 四個圖
#plt.subplot(2,3,1)
#plt.text(0.5,0.5,'No Pic')
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), '-y', label='sin(x)')
plt.legend(loc='lower right', frameon=True, ncol=2)
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), '-c', label='cos(x)')
plt.legend(loc='lower right', frameon=True, ncol=2)
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), '-g', label='tan(x)')
plt.legend(loc='lower right', frameon=True, ncol=2)
plt.subplot(2, 2, 4)
plt.plot(x, np.tanh(x), '-b', label='tanh(x)')
plt.legend(loc='lower right', frameon=True, ncol=2)
plt.show()
```

### 使用不等大小子圖的技巧

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
# o -> 圖示樣式 k -> 圖示顏色
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
