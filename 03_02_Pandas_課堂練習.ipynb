```python
#找出總共有多少種不同的平台
data['Platform'].unique()
```

```python
differ = data['Platform'].unique()
print(differ)
print("總共有", differ.size, "種不同的平台")
```

```python
#寫程式計算不同平台的總銷售量
differ = data['Platform'].unique()
for Name_Platform in differ:
    Sales_total = data[data["Platform"] == Name_Platform]["Global_Sales"].sum()
    print(Name_Platform, ":", Sales_total)
```

```python
# 使用GroupBy來完成上述的工作
data.groupby("Platform")["Global_Sales"].sum()
```
