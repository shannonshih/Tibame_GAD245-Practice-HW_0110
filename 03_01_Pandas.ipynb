# DataFrame
二維數據結構，類似於 Excel 表格或 SQL 數據表，是 Pandas 中最常用的數據結構。它由多個 Series 組成，每個 Series 代表一列。

```python
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)

area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)
area
```

```python
states = pd.DataFrame({'人口': population,
                       '面積': area})
states
```

```python
states.index
```

```python
states.columns
```

```python
# 練習題
# 請建立另一個體重與姓名關係的Series
# 並再建立一個包含 體重與身高 的 DataFrame,使用姓名當索引
height_dict = {'John':185, 'Fernandes':172, 'Stephen':190, 'Shannon':164, 'Alexender':220, 'Julia':153, 'Sandra':178}
height = pd.Series(height_dict)
print(height)

weight_dict = {'John':80, 'Fernandes':65, 'Stephen':72, 'Shannon':75, 'Alexender':100, 'Julia':55, 'Sandra':62}
weight = pd.Series(weight_dict)
print(weight)

person_information = pd.DataFrame({'身高': height,
                       '體重': weight})

print(person_information)
```

#### 練習題用資料

```python
h = {'David':162, 'Ken':169, 'Sam': 173, 'Owen':165, 'Paul':177}
w = {'David':70, 'Owen':80, 'Paul':78, 'Ken':65, 'Sam': 83}
person = pd.DataFrame({'身高': h, '體重': w})
person
```

```python
#練習題
#請列出第 1~3 筆資料
person.iloc[0:3]
```

```python
# 列出第 2,4,5筆資料
# 下面的寫法為何失敗？
person.loc[[1,3,4]]
```

```python
person.iloc[[1,3,4]]
```

```python
person.loc[['Ken','Owen','Paul']]
```

### 問題： loc 跟 iloc 有什麼不同？

```python
person['身高']
```

```python
person.身高
```

```python
# 練習題
# 請算出每個人的 BMI 指數
person.體重 / ((person.身高/100)**2)
```

```python
# 練習題
# 猜猜如何將算出來的 BMI 變成新欄位併入原始資料表中
# 方法1
bmi = person.體重 / ((person.身高/100)**2)
person = pd.DataFrame({'身高': h, '體重': w, 'BMI': bmi})
print(person)
print("-" * 30)

# 方法2
person['BMI'] = person.體重 / ((person.身高/100)**2)
print(person)
```
